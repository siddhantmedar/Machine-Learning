```python
import numpy as np
from mlxtend.plotting import plot_decision_regions

from sklearn.linear_model import Perceptron
```

```python
total_data = 150
train_data = 100
```

```python
X = np.random.uniform(0,1,(total_data,2))
y = np.random.randint(0,2,total_data)

X.shape, y.shape
```

```python
model = Perceptron()
```

```python
model.fit(X[:train_data],y[:train_data])
```

```python
y_true = y[train_data:]
y_pred = model.predict(X[train_data:])
```

```python
(y_true==y_pred).sum() / y[train_data:].shape[0]
```

```python
model.coef_
```

```python
plot_decision_regions(X,y,clf=model,legend=2)
```
