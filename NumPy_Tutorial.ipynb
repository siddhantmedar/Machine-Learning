#### Broadcasting

Broadcasting is the technique of modifying two arrays of unequal sizes such that a binary
ufunc can be applied between them.

Broadcasting is possible between two arrays of different shapes only if one of the following
conditions are satisfied:

a. One of the arrays is a single element array.<br>
b. Both arrays have one non similar axis with the same size.

If the above conditions are not met during a binary operation, numpy raises an error.

```python
lst2 = np.random.randn(5)
lst1, lst2
```

```python
# lst1+lst2 this fails without broadcasting
```

Two norm types are relevant for our purposes, they are the
L1 and L2 norms. The L1 norm is the sum of the absolute values of each element of the vector,
whereas the L2 norm is the square root of the sum of the squares of each element in the vector.

norm without axis consider diagonal elements
and axis-0 and axis-1 is understood.

```python
np.linalg.norm(lst,1),np.linalg.norm(lst, 1, axis=0), np.linalg.norm(lst, 1, axis=1)
```

```python
np.linalg.norm(lst,2),np.linalg.norm(lst, 2, axis=0), np.linalg.norm(lst, 2, axis=1)
```

### Dot Product

The dot product is the most fundamental linear algebraic operation between two equal sized
vectors (ie: one dimensional array). The dot product can be deconstructed into two sequential <br>
operations:<br>
1. Multiplication operation between the two one dimensional arrays.<br>
2. Summation of the values of the vector resulting from step 1.

```python
lst1, lst2
```

```python
lst1 = np.random.randn(3)
lst2 = np.random.randn(3)

lst1, lst2
```

```python
np.sum((np.multiply(lst1, lst2))), np.dot(lst1, lst2), lst1@lst2
```
